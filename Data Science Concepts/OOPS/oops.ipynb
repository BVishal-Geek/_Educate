```python
import pandas as pd

Df = pd.read_csv('Automobile.csv')
Df.describe()
```

```python
'''
Code Explanation: 

I am trying to read car csv and trying to segregate all the details according to the information. 

I am iterating all the rows into the class and calculating ACDH values i.e. Accelaration x Cylinders x Displacement x Horsepower, and we are storing 
that values in a new column called ACDH. 

'''

class Car: 
    class Details:
        def __init__(self, row): # This row is attribute
            ''' 
            __init__ is the constructor. For example, we have a 300 row dataframe and we are pushing each row continuously. All the row values are
            updated to the parameters automatically. That is the reason we use constructors.  
            All the variables with the self is called `Parameter`
            '''
            self.name = row['name']
            self.mpg = row['mpg']
            self.cylinders = row['cylinders']
            self.displacement = row['displacement']
            self.horsepower = row['horsepower']
            self.weight = row['weight']
            self.acceleration = row['acceleration']
            self.model_year = row['model_year']
            self.origin = row['origin']
            # return self.name, self.mpg, self.cylinders, self.displacement, self.horsepower, self.weight, self.acceleration, self.model_year, self.origin

    class EngineDetails:
        def __init__(self, row):
            self.mpg = row['mpg'].astype(int)
            self.cylinders = row['cylinders'].astype(int)
            self.displacement = row['displacement'].astype(int)
            self.horsepower = row['horsepower'].astype(int)
            self.acceleration = row['acceleration'].astype(float)

        def calculation(self):
            return self.cylinders * self.displacement * self.horsepower * self.acceleration
          
    class ModelDetails:
        def __init__(self, row):
            self.name = row['name']
            self.model_year = row['model_year']
            self.origin = row['origin']

car_ = Car()

Df.apply(lambda row: car_.Details(row), axis= 1)  # row is argument 
Df['ACDH'] = Df.apply(lambda row: car_.EngineDetails.calculation(row), axis= 1) # Created a new column and updating all the values to the new columnn. 



```
